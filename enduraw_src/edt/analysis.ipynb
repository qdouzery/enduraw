## Imports

```python
import polars as pl
import plotly.graph_objects as go
```

## Data

```python
df_results = pl.read_parquet(
    "/Users/qdouzery/Desktop/enduraw/data/edt-2024_results/df_results.parquet"
)
```

## Exploratory analysis

### Finish times

```python
df_results_pp = df_results.with_columns(
    pl.col("time").cast(pl.Duration()).dt.total_seconds().alias("time_s")
)
```

```python
fig = go.Figure()
fig.add_trace(
    go.Histogram(
        x=df_results_pp["time_s"].to_numpy(),
        histnorm="percent",
        nbinsx=100,
        marker_color="gold",
        showlegend=False,
    )
)
fig.update_xaxes(title_text="<b>Finish time (s)</b>")
fig.update_yaxes(title_text="<b>Percentage (%)</b>")
fig.update_layout(
    title_text="<b>Finish times distribution</b><br>All athletes",
    title_x=0.5,
    title_xanchor="center",
    font=dict(size=12),
    plot_bgcolor="white",  # paper_bgcolor='gainsboro'
    xaxis={"showgrid": True, "gridcolor": "lightgrey", "gridwidth": 0.5},
    yaxis={"showgrid": True, "gridcolor": "lightgrey", "gridwidth": 0.5},
)
fig.update_traces(marker_line_width=1, marker_line_color="black")
fig.show()
```
